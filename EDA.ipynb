What techniques can be used to handle missing data?

techniques used to handle missing and null data in tha dataset given are:
1.Measure of central tendencies: This can be applied based on the data we have
a.Mean:Mean is the avg of the columns and can be filled in the place of null values
b.Median:median gives the centre value of whole data provided that can be replaced in place of missing values
c.Mode:mode gives us the most occured value in tha data that can be used to replace with the missing values in the dataset
2. We can drop the missing values if they are not having much importance in the dataset and thus does not effect the inferences



```python
# Example on Measure of tendencies 
s=[[1,2,3,4,np.nan],[np.nan,43,4,2],[1,2,4,5]]
pd.Series(s)
s=pd.DataFrame(s)
s
```

```python
s.isnull().sum().sum()
```

```python
s[0].mode() #applying mode on s[0] column
```

```python
s[0].replace(np.nan,1) #replacing the missing value with the mode value.
```

```python
# since the last column has all nan values that row is not needed so we can drop the column
s.drop(4,axis=1)
```
